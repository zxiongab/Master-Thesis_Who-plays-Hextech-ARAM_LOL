```python
"""
League of Legends ARAM: Mayhem 分析脚本
=============================================
复刻论文 §3.2.1 的分层随机采样方法，分析各 Tier 玩家的
ARAM: Mayhem 参与比例，以及 2025.10.21 前后每人单位时间内
LoL 游戏场次的变化（ARAM: Mayhem 于 2025-10-22 上线）。

使用方式：
    pip install requests pandas scipy tqdm
    python riot_aram_mayhem_analysis.py --api-key <YOUR_RIOT_API_KEY> --region euw1

参数：
    --api-key       Riot 开发者 API Key（可在 https://developer.riotgames.com 获取）
    --region        平台区域，默认 euw1（可选: na1, kr, euw1, eune1, jp1 等）
    --regional      区域路由，默认 europe（与 region 对应: americas/asia/europe/sea）
    --samples       每个 sub-tier 采样玩家数，默认 100（论文用 1000，受 rate limit 限制）
    --matches       每位玩家抓取的最近比赛数，默认 20（论文相同）
    --output-dir    输出目录，默认 ./output
"""

import argparse
import json
import os
import time
import random
import logging
from datetime import datetime, timezone
from collections import defaultdict

import requests
import pandas as pd
from scipy import stats
from tqdm import tqdm

# ──────────────────────────────────────────────
# 配置
# ──────────────────────────────────────────────

# ARAM: Mayhem 上线时间（UTC）
MAYHEM_LAUNCH_TS = int(datetime(2025, 10, 22, tzinfo=timezone.utc).timestamp() * 1000)
MAYHEM_LAUNCH_DATE = "2025-10-22"

# 论文复刻：Iron → Diamond（排除 Master/GM/Challenger）
TIERS = ["IRON", "BRONZE", "SILVER", "GOLD", "PLATINUM", "EMERALD", "DIAMOND"]
DIVISIONS = ["IV", "III", "II", "I"]  # 每个 Tier 的 4 个 sub-tier

# Queue types
QUEUE_RANKED = "RANKED_SOLO_5x5"
QUEUE_ARAM = "ARAM"           # queue id 450
QUEUE_MAYHEM = "CHERRY"        # ARAM: Mayhem 的 queue id（实际 id 需确认，暂用 CHERRY）

# queue_id → mode 名称
QUEUE_ID_MAP = {
    400: "Draft Pick",
    420: "Ranked Solo/Duo",
    430: "Blind Pick",
    440: "Ranked Flex",
    450: "ARAM",
    700: "Clash",
    720: "ARAM Clash",
    830: "Co-op vs AI",
    900: "URF",
    1300: "Nexus Blitz",
    1400: "Ultimate Spellbook",
    1700: "Arena",
    1900: "URF",
    2000: "Tutorial",
    2010: "Tutorial",
    2020: "Tutorial",
    1810: "ARAM: Mayhem",   # 推测 id，需根据实际 API 返回确认
}

# 已知 ARAM Mayhem 可能的 queue id（Riot 有时用不同 id）
MAYHEM_QUEUE_IDS = {1810, 1820}  # 上线后需从实际数据确认

# ──────────────────────────────────────────────
# Rate limit 处理（开发者 key: 20 req/s, 100 req/2min）
# ──────────────────────────────────────────────

logging.basicConfig(level=logging.INFO, format="%(asctime)s %(levelname)s %(message)s")
logger = logging.getLogger(__name__)


class RiotClient:
    def __init__(self, api_key: str, region: str = "euw1", regional: str = "europe"):
        self.api_key = api_key
        self.region = region        # platform: na1, euw1, kr, ...
        self.regional = regional    # regional: americas, europe, asia, sea
        self.session = requests.Session()
        self.session.headers.update({"X-Riot-Token": api_key})
        self._call_times = []

    def _rate_limit(self):
        """简单的速率限制：保持每秒 ≤ 18 次请求"""
        now = time.time()
        self._call_times = [t for t in self._call_times if now - t < 1.0]
        if len(self._call_times) >= 18:
            time.sleep(1.05 - (now - self._call_times[0]))
        self._call_times.append(time.time())

    def _get(self, url: str, retries: int = 5) -> dict:
        for attempt in range(retries):
            self._rate_limit()
            try:
                resp = self.session.get(url, timeout=10)
                if resp.status_code == 200:
                    return resp.json()
                elif resp.status_code == 429:
                    retry_after = int(resp.headers.get("Retry-After", 10))
                    logger.warning(f"Rate limited. Sleeping {retry_after}s...")
                    time.sleep(retry_after + 1)
                elif resp.status_code == 404:
                    return None
                else:
                    logger.warning(f"HTTP {resp.status_code} for {url}. Attempt {attempt+1}/{retries}")
                    time.sleep(2 ** attempt)
            except Exception as e:
                logger.warning(f"Request error: {e}. Attempt {attempt+1}/{retries}")
                time.sleep(2 ** attempt)
        return None

    # ── League-V4 ──────────────────────────────
    def get_league_entries(self, tier: str, division: str, page: int = 1) -> list:
        url = (f"https://{self.region}.api.riotgames.com/lol/league/v4/entries"
               f"/{QUEUE_RANKED}/{tier}/{division}?page={page}")
        result = self._get(url)
        return result or []

    def get_summoner_by_id(self, summoner_id: str) -> dict:
        url = f"https://{self.region}.api.riotgames.com/lol/summoner/v4/summoners/{summoner_id}"
        return self._get(url)

    # ── Account-V1 (puuid) ─────────────────────
    def get_account_by_puuid(self, puuid: str) -> dict:
        url = f"https://{self.regional}.api.riotgames.com/riot/account/v1/accounts/by-puuid/{puuid}"
        return self._get(url)

    # ── Match-V5 ──────────────────────────────
    def get_match_ids(self, puuid: str, count: int = 20,
                      start_time: int = None, end_time: int = None) -> list:
        params = f"count={count}"
        if start_time:
            params += f"&startTime={start_time // 1000}"  # match-v5 用秒
        if end_time:
            params += f"&endTime={end_time // 1000}"
        url = (f"https://{self.regional}.api.riotgames.com/lol/match/v5/matches"
               f"/by-puuid/{puuid}/ids?{params}")
        result = self._get(url)
        return result or []

    def get_match(self, match_id: str) -> dict:
        url = f"https://{self.regional}.api.riotgames.com/lol/match/v5/matches/{match_id}"
        return self._get(url)


# ──────────────────────────────────────────────
# 采样：复刻论文分层随机采样
# ──────────────────────────────────────────────

def sample_players(client: RiotClient, samples_per_subtier: int = 100) -> pd.DataFrame:
    """
    从每个 Tier×Division（共 7×4=28 个 sub-tier）随机采样 samples_per_subtier 名玩家。
    论文用 1000，默认用 100（受开发者 key rate limit 限制）。
    总样本量 = 28 × samples_per_subtier。
    """
    rows = []
    for tier in TIERS:
        for div in DIVISIONS:
            logger.info(f"Sampling {tier} {div}...")
            # 先获取第 1 页，了解大概有多少玩家
            all_entries = []
            for page in range(1, 6):  # 最多抓 5 页（每页 205 条）
                entries = client.get_league_entries(tier, div, page)
                if not entries:
                    break
                all_entries.extend(entries)
                if len(entries) < 205:
                    break  # 已是最后一页

            if not all_entries:
                logger.warning(f"No entries found for {tier} {div}")
                continue

            # 随机采样
            sampled = random.sample(all_entries, min(samples_per_subtier, len(all_entries)))

            for entry in sampled:
                rows.append({
                    "summonerId": entry["summonerId"],
                    "tier": tier,
                    "division": div,
                    "tier_division": f"{tier} {div}",
                    "leaguePoints": entry.get("leaguePoints", 0),
                    "wins": entry.get("wins", 0),
                    "losses": entry.get("losses", 0),
                })

    df = pd.DataFrame(rows)
    logger.info(f"Total sampled players: {len(df)}")
    return df


def enrich_with_puuid(client: RiotClient, df: pd.DataFrame) -> pd.DataFrame:
    """通过 summonerId → puuid"""
    puuids = []
    for sid in tqdm(df["summonerId"], desc="Fetching PUUIDs"):
        summoner = client.get_summoner_by_id(sid)
        puuids.append(summoner["puuid"] if summoner else None)
    df["puuid"] = puuids
    return df.dropna(subset=["puuid"])


# ──────────────────────────────────────────────
# 比赛数据抓取
# ──────────────────────────────────────────────

def fetch_player_matches(client: RiotClient, puuid: str, count: int = 20) -> list:
    """抓取最近 count 场比赛（不限时间，用于 Mayhem 参与率分析）"""
    match_ids = client.get_match_ids(puuid, count=count)
    matches = []
    for mid in match_ids:
        m = client.get_match(mid)
        if m:
            info = m.get("info", {})
            matches.append({
                "matchId": mid,
                "queueId": info.get("queueId"),
                "gameMode": info.get("gameMode"),
                "gameName": info.get("gameName", ""),
                "gameStartTimestamp": info.get("gameStartTimestamp"),
                "gameDuration": info.get("gameDuration"),
            })
    return matches


def is_mayhem(match: dict) -> bool:
    """判断是否为 ARAM: Mayhem"""
    qid = match.get("queueId")
    if qid in MAYHEM_QUEUE_IDS:
        return True
    # 备用：检查 gameMode 或 gameName
    name = (match.get("gameName") or "").lower()
    if "mayhem" in name or "aramm" in name:
        return True
    return False


# ──────────────────────────────────────────────
# 分析1：各 Tier 的 Mayhem 参与比例
# ──────────────────────────────────────────────

def analyze_mayhem_ratio(player_df: pd.DataFrame, all_matches: dict) -> pd.DataFrame:
    """
    计算每个 Tier 中玩过至少 1 场 ARAM Mayhem 的玩家比例，
    以及整体 Mayhem 场次占所有场次的比例。
    """
    results = []
    for tier in TIERS:
        tier_players = player_df[player_df["tier"] == tier]["puuid"].tolist()
        played_mayhem = 0
        total_mayhem_games = 0
        total_games = 0

        for puuid in tier_players:
            matches = all_matches.get(puuid, [])
            total_games += len(matches)
            mayhem_count = sum(1 for m in matches if is_mayhem(m))
            total_mayhem_games += mayhem_count
            if mayhem_count > 0:
                played_mayhem += 1

        n = len(tier_players)
        results.append({
            "tier": tier,
            "n_players": n,
            "players_played_mayhem": played_mayhem,
            "player_mayhem_ratio": played_mayhem / n if n > 0 else 0,
            "total_games": total_games,
            "total_mayhem_games": total_mayhem_games,
            "game_mayhem_ratio": total_mayhem_games / total_games if total_games > 0 else 0,
        })

    # 整体
    all_puuids = player_df["puuid"].tolist()
    total_p_mayhem = sum(
        1 for p in all_puuids
        if any(is_mayhem(m) for m in all_matches.get(p, []))
    )
    total_g = sum(len(all_matches.get(p, [])) for p in all_puuids)
    total_mg = sum(
        sum(1 for m in all_matches.get(p, []) if is_mayhem(m))
        for p in all_puuids
    )
    results.append({
        "tier": "OVERALL",
        "n_players": len(all_puuids),
        "players_played_mayhem": total_p_mayhem,
        "player_mayhem_ratio": total_p_mayhem / len(all_puuids) if all_puuids else 0,
        "total_games": total_g,
        "total_mayhem_games": total_mg,
        "game_mayhem_ratio": total_mg / total_g if total_g > 0 else 0,
    })

    return pd.DataFrame(results)


# ──────────────────────────────────────────────
# 分析2：Mayhem 上线前后每人单位时间内场次变化
# 时间窗口：上线前 30 天 vs 上线后 30 天
# ──────────────────────────────────────────────

WINDOW_DAYS = 30
SECS_IN_DAY = 86400
MS_IN_DAY = SECS_IN_DAY * 1000

# 时间窗口（毫秒）
PRE_END_TS = MAYHEM_LAUNCH_TS                              # 2025-10-22 00:00 UTC
PRE_START_TS = PRE_END_TS - WINDOW_DAYS * MS_IN_DAY        # 2025-09-22

POST_START_TS = MAYHEM_LAUNCH_TS                           # 2025-10-22 00:00 UTC
POST_END_TS = POST_START_TS + WINDOW_DAYS * MS_IN_DAY      # 2025-11-21


def fetch_windowed_matches(client: RiotClient, puuid: str) -> dict:
    """
    分别抓取 Mayhem 上线前 30 天 / 后 30 天的比赛数
    返回 {'pre': [matches], 'post': [matches]}
    """
    pre_ids = client.get_match_ids(
        puuid, count=100,
        start_time=PRE_START_TS, end_time=PRE_END_TS
    )
    post_ids = client.get_match_ids(
        puuid, count=100,
        start_time=POST_START_TS, end_time=POST_END_TS
    )
    # 注意：此处只取 match id 数量，不抓详情（节省 API 配额）
    return {"pre_count": len(pre_ids), "post_count": len(post_ids)}


def analyze_activity_change(player_df: pd.DataFrame,
                             windowed_data: dict) -> pd.DataFrame:
    """
    对每个 Tier 比较上线前后 30 天内每人平均场次，
    并用配对 t 检验检验差异显著性。
    """
    results = []
    all_pre, all_post = [], []

    for tier in TIERS:
        tier_puuids = player_df[player_df["tier"] == tier]["puuid"].tolist()
        pre_counts = [windowed_data[p]["pre_count"] for p in tier_puuids if p in windowed_data]
        post_counts = [windowed_data[p]["post_count"] for p in tier_puuids if p in windowed_data]

        n = len(pre_counts)
        if n < 2:
            continue

        all_pre.extend(pre_counts)
        all_post.extend(post_counts)

        mean_pre = sum(pre_counts) / n
        mean_post = sum(post_counts) / n
        t_stat, p_val = stats.ttest_rel(post_counts, pre_counts)

        results.append({
            "tier": tier,
            "n": n,
            "window_days": WINDOW_DAYS,
            "mean_games_pre_30d": round(mean_pre, 2),
            "mean_games_post_30d": round(mean_post, 2),
            "mean_games_per_day_pre": round(mean_pre / WINDOW_DAYS, 3),
            "mean_games_per_day_post": round(mean_post / WINDOW_DAYS, 3),
            "change_abs": round(mean_post - mean_pre, 2),
            "change_pct": round((mean_post - mean_pre) / mean_pre * 100, 1) if mean_pre > 0 else None,
            "t_stat": round(t_stat, 3),
            "p_value": round(p_val, 4),
            "significant_p05": p_val < 0.05,
        })

    # Overall
    n_all = len(all_pre)
    if n_all >= 2:
        m_pre = sum(all_pre) / n_all
        m_post = sum(all_post) / n_all
        t_stat, p_val = stats.ttest_rel(all_post, all_pre)
        results.append({
            "tier": "OVERALL",
            "n": n_all,
            "window_days": WINDOW_DAYS,
            "mean_games_pre_30d": round(m_pre, 2),
            "mean_games_post_30d": round(m_post, 2),
            "mean_games_per_day_pre": round(m_pre / WINDOW_DAYS, 3),
            "mean_games_per_day_post": round(m_post / WINDOW_DAYS, 3),
            "change_abs": round(m_post - m_pre, 2),
            "change_pct": round((m_post - m_pre) / m_pre * 100, 1) if m_pre > 0 else None,
            "t_stat": round(t_stat, 3),
            "p_value": round(p_val, 4),
            "significant_p05": p_val < 0.05,
        })

    return pd.DataFrame(results)


# ──────────────────────────────────────────────
# 主流程
# ──────────────────────────────────────────────

def main():
    parser = argparse.ArgumentParser(description="Riot API ARAM Mayhem Analysis")
    parser.add_argument("--api-key", required=True, help="Riot developer API key")
    parser.add_argument("--region", default="euw1",
                        help="Platform region (na1/euw1/kr/eune1/jp1/br1/la1/la2/oc1/tr1/ru)")
    parser.add_argument("--regional", default="europe",
                        help="Regional routing (americas/europe/asia/sea)")
    parser.add_argument("--samples", type=int, default=100,
                        help="Players sampled per sub-tier (paper uses 1000)")
    parser.add_argument("--matches", type=int, default=20,
                        help="Recent matches per player for Mayhem ratio analysis")
    parser.add_argument("--output-dir", default="./output")
    args = parser.parse_args()

    os.makedirs(args.output_dir, exist_ok=True)
    client = RiotClient(args.api_key, args.region, args.regional)

    # ── Step 1: 分层采样 ──────────────────────
    logger.info("=" * 50)
    logger.info("Step 1: Stratified random sampling (replicating paper §3.2.1)")
    logger.info(f"  Tiers: {TIERS}")
    logger.info(f"  Sub-tiers per tier: {len(DIVISIONS)}")
    logger.info(f"  Samples per sub-tier: {args.samples}")
    logger.info(f"  Expected total: {len(TIERS) * len(DIVISIONS) * args.samples}")
    logger.info("=" * 50)

    players_path = os.path.join(args.output_dir, "sampled_players.csv")
    if os.path.exists(players_path):
        logger.info("Loading cached player sample...")
        player_df = pd.read_csv(players_path)
    else:
        player_df = sample_players(client, args.samples)
        player_df = enrich_with_puuid(client, player_df)
        player_df.to_csv(players_path, index=False)
        logger.info(f"Saved {len(player_df)} players to {players_path}")

    logger.info(f"Sample breakdown:\n{player_df.groupby('tier').size()}")

    # ── Step 2: 抓取最近比赛（用于 Mayhem 参与率）──
    logger.info("Step 2: Fetching recent matches for Mayhem participation analysis...")
    matches_path = os.path.join(args.output_dir, "all_matches.json")
    if os.path.exists(matches_path):
        with open(matches_path) as f:
            all_matches = json.load(f)
    else:
        all_matches = {}
        for puuid in tqdm(player_df["puuid"].tolist(), desc="Fetching matches"):
            all_matches[puuid] = fetch_player_matches(client, puuid, args.matches)
        with open(matches_path, "w") as f:
            json.dump(all_matches, f)

    # 自动检测 Mayhem queue id（从实际数据中发现）
    detected_qids = set()
    for matches in all_matches.values():
        for m in matches:
            qid = m.get("queueId")
            name = (m.get("gameName") or "").lower()
            if "mayhem" in name or "aramm" in name:
                detected_qids.add(qid)
                MAYHEM_QUEUE_IDS.add(qid)
    if detected_qids:
        logger.info(f"Detected ARAM Mayhem queue IDs from data: {detected_qids}")
    else:
        logger.warning("No ARAM Mayhem games detected in sampled matches. "
                       "Try increasing --matches or check if the mode is active.")

    # ── Step 3: 分析 Mayhem 参与率 ────────────
    logger.info("Step 3: Analyzing ARAM Mayhem participation ratio per tier...")
    ratio_df = analyze_mayhem_ratio(player_df, all_matches)
    ratio_path = os.path.join(args.output_dir, "mayhem_ratio_by_tier.csv")
    ratio_df.to_csv(ratio_path, index=False)

    print("\n" + "=" * 60)
    print("📊 ARAM: Mayhem 参与率（按 Tier）")
    print("=" * 60)
    for _, row in ratio_df.iterrows():
        print(f"  {row['tier']:10s} | "
              f"玩家参与率: {row['player_mayhem_ratio']:.1%} "
              f"({row['players_played_mayhem']}/{row['n_players']}) | "
              f"场次占比: {row['game_mayhem_ratio']:.1%} "
              f"({row['total_mayhem_games']}/{row['total_games']})")

    # ── Step 4: 抓取前后 30 天窗口数据 ────────
    logger.info("Step 4: Fetching pre/post Mayhem launch windowed match counts...")
    logger.info(f"  Pre window:  {datetime.utcfromtimestamp(PRE_START_TS/1000).date()} → "
                f"{datetime.utcfromtimestamp(PRE_END_TS/1000).date()}")
    logger.info(f"  Post window: {datetime.utcfromtimestamp(POST_START_TS/1000).date()} → "
                f"{datetime.utcfromtimestamp(POST_END_TS/1000).date()}")

    windowed_path = os.path.join(args.output_dir, "windowed_counts.json")
    if os.path.exists(windowed_path):
        with open(windowed_path) as f:
            windowed_data = json.load(f)
    else:
        windowed_data = {}
        for puuid in tqdm(player_df["puuid"].tolist(), desc="Windowed match counts"):
            windowed_data[puuid] = fetch_windowed_matches(client, puuid)
        with open(windowed_path, "w") as f:
            json.dump(windowed_data, f)

    # ── Step 5: 活跃度变化分析 ────────────────
    logger.info("Step 5: Analyzing activity change (games/day pre vs post launch)...")
    activity_df = analyze_activity_change(player_df, windowed_data)
    activity_path = os.path.join(args.output_dir, "activity_change_by_tier.csv")
    activity_df.to_csv(activity_path, index=False)

    print("\n" + "=" * 70)
    print(f"📈 ARAM Mayhem 上线前后 {WINDOW_DAYS} 天内每人平均 LoL 场次变化")
    print(f"   前: {datetime.utcfromtimestamp(PRE_START_TS/1000).date()} ~ "
          f"{datetime.utcfromtimestamp(PRE_END_TS/1000).date()}")
    print(f"   后: {datetime.utcfromtimestamp(POST_START_TS/1000).date()} ~ "
          f"{datetime.utcfromtimestamp(POST_END_TS/1000).date()}")
    print("=" * 70)
    for _, row in activity_df.iterrows():
        sig = "✅ 显著" if row["significant_p05"] else "❌ 不显著"
        print(f"  {row['tier']:10s} | "
              f"前:{row['mean_games_per_day_pre']:.3f} → 后:{row['mean_games_per_day_post']:.3f} "
              f"场/天 | {row['change_pct']:+.1f}% | "
              f"t={row['t_stat']:.3f}, p={row['p_value']:.4f} {sig}")

    print(f"\n所有结果已保存至 {args.output_dir}/")
    print("  sampled_players.csv       - 分层采样玩家数据")
    print("  all_matches.json          - 最近比赛原始数据")
    print("  mayhem_ratio_by_tier.csv  - 各 Tier Mayhem 参与率")
    print("  activity_change_by_tier.csv - 上线前后活跃度对比")


if __name__ == "__main__":
    main()
```

```python
# ============================================================
# LoL ARAM: Mayhem 分析 MVP
# 在 Google Colab 中直接运行
# ============================================================

# ── Cell 1: 安装依赖 ─────────────────────────────────────
!pip install requests pandas scipy tqdm -q

# ── Cell 2: 设置 API Key ─────────────────────────────────
from google.colab import userdata
import os

# 如果你用了 Colab Secrets（推荐）：
#os.environ["RIOT_API_KEY"] = userdata.get("RIOT_API_KEY")

# 如果没有设置 Secrets，直接粘贴 key（不推荐）：
os.environ["RIOT_API_KEY"] = "RGAPI-e9aa2901-9fce-4599-a271-b7ce7b927162"

API_KEY = os.environ["RIOT_API_KEY"]
print("✅ API Key loaded" if API_KEY else "❌ API Key missing")

# ── Cell 3: 配置参数 ─────────────────────────────────────
# MVP 参数：小样本，快速验证
CONFIG = {
    "region":   "kr",      # 平台路由
    "regional": "asion",    # 区域路由
    "samples":  10,          # 每个 sub-tier 采样人数（MVP 用 10，正式用 100-1000）
    "matches":  20,          # 每人抓取最近比赛数
}

TIERS     = ["IRON", "BRONZE", "SILVER", "GOLD", "PLATINUM", "EMERALD", "DIAMOND"]
DIVISIONS = ["IV", "III", "II", "I"]

from datetime import datetime, timezone
MAYHEM_LAUNCH_TS = int(datetime(2025, 10, 22, tzinfo=timezone.utc).timestamp() * 1000)
WINDOW_DAYS = 30
MS_IN_DAY   = 86400 * 1000

PRE_START_TS  = MAYHEM_LAUNCH_TS - WINDOW_DAYS * MS_IN_DAY
PRE_END_TS    = MAYHEM_LAUNCH_TS
POST_START_TS = MAYHEM_LAUNCH_TS
POST_END_TS   = MAYHEM_LAUNCH_TS + WINDOW_DAYS * MS_IN_DAY

print(f"Pre  window: {datetime.utcfromtimestamp(PRE_START_TS/1000).date()} → {datetime.utcfromtimestamp(PRE_END_TS/1000).date()}")
print(f"Post window: {datetime.utcfromtimestamp(POST_START_TS/1000).date()} → {datetime.utcfromtimestamp(POST_END_TS/1000).date()}")

# ── Cell 4: Riot API Client ──────────────────────────────
import requests, time, random, logging
logging.basicConfig(level=logging.WARNING)

class RiotClient:
    def __init__(self, api_key, region, regional):
        self.api_key  = api_key
        self.region   = region
        self.regional = regional
        self.session  = requests.Session()
        self.session.headers.update({"X-Riot-Token": api_key})
        self._calls   = []

    def _throttle(self):
        now = time.time()
        self._calls = [t for t in self._calls if now - t < 1.2]
        if len(self._calls) >= 18:
            time.sleep(1.2 - (now - self._calls[0]))
        self._calls.append(time.time())

    def _get(self, url, retries=5):
        for i in range(retries):
            self._throttle()
            try:
                r = self.session.get(url, timeout=10)
                if r.status_code == 200:
                    return r.json()
                elif r.status_code == 429:
                    wait = int(r.headers.get("Retry-After", 15))
                    print(f"⏳ Rate limited, waiting {wait}s...")
                    time.sleep(wait + 1)
                elif r.status_code == 404:
                    return None
                else:
                    print(f"⚠️  HTTP {r.status_code}: {url[:80]}")
                    time.sleep(2 ** i)
            except Exception as e:
                print(f"⚠️  Error: {e}")
                time.sleep(2 ** i)
        return None

    def get_league_entries(self, tier, division, page=1):
        url = (f"https://{self.region}.api.riotgames.com/lol/league/v4/entries"
               f"/RANKED_SOLO_5x5/{tier}/{division}?page={page}")
        return self._get(url) or []

    def get_summoner(self, summoner_id):
        url = f"https://{self.region}.api.riotgames.com/lol/summoner/v4/summoners/{summoner_id}"
        return self._get(url)

    def get_match_ids(self, puuid, count=20, start_ts=None, end_ts=None):
        params = f"count={count}"
        if start_ts: params += f"&startTime={start_ts // 1000}"
        if end_ts:   params += f"&endTime={end_ts // 1000}"
        url = (f"https://{self.regional}.api.riotgames.com"
               f"/lol/match/v5/matches/by-puuid/{puuid}/ids?{params}")
        return self._get(url) or []

    def get_match(self, match_id):
        url = f"https://{self.regional}.api.riotgames.com/lol/match/v5/matches/{match_id}"
        return self._get(url)

client = RiotClient(API_KEY, CONFIG["region"], CONFIG["regional"])
print("✅ Client ready")

# ── Cell 5: 分层采样 ─────────────────────────────────────
import pandas as pd
from tqdm.notebook import tqdm

def sample_players(client, samples_per_subtier):
    rows = []
    for tier in tqdm(TIERS, desc="Tiers"):
        for div in DIVISIONS:
            entries = []
            for page in range(1, 4):
                batch = client.get_league_entries(tier, div, page)
                if not batch: break
                entries.extend(batch)
                if len(batch) < 205: break

            if not entries:
                print(f"⚠️  No entries: {tier} {div}")
                continue

            sampled = random.sample(entries, min(samples_per_subtier, len(entries)))
            for e in sampled:
                rows.append({
                    "summonerId":    e["summonerId"],
                    "tier":          tier,
                    "division":      div,
                    "leaguePoints":  e.get("leaguePoints", 0),
                })
    # 在 sample_players 函数里，sampled 之后加这两行调试
    print("Sample entry keys:", sampled[0].keys())
    print("Sample entry:", sampled[0])
    return pd.DataFrame(rows)

print(f"Sampling {CONFIG['samples']} players per sub-tier "
      f"({len(TIERS)} tiers × {len(DIVISIONS)} divs = "
      f"{len(TIERS)*len(DIVISIONS)*CONFIG['samples']} total expected)...")

player_df = sample_players(client, CONFIG["samples"])
print(f"✅ Sampled {len(player_df)} players")
print(player_df.groupby("tier").size().to_string())

# ── Cell 6: 获取 PUUID ───────────────────────────────────
puuids = []
for sid in tqdm(player_df["summonerId"], desc="Fetching PUUIDs"):
    s = client.get_summoner(sid)
    puuids.append(s["puuid"] if s else None)

player_df["puuid"] = puuids
player_df = player_df.dropna(subset=["puuid"])
print(f"✅ Got PUUIDs for {len(player_df)} players")

# ── Cell 7: 抓最近比赛（Mayhem 参与率用）───────────────
MAYHEM_QUEUE_IDS = {1810, 1820}  # 上线后从数据自动更新

def is_mayhem(match_info):
    if match_info.get("queueId") in MAYHEM_QUEUE_IDS:
        return True
    name = (match_info.get("gameName") or "").lower()
    return "mayhem" in name

all_matches = {}  # puuid → [match_info, ...]

for puuid in tqdm(player_df["puuid"], desc="Fetching recent matches"):
    ids = client.get_match_ids(puuid, count=CONFIG["matches"])
    matches = []
    for mid in ids:
        m = client.get_match(mid)
        if m:
            info = m.get("info", {})
            matches.append({
                "matchId":             mid,
                "queueId":             info.get("queueId"),
                "gameMode":            info.get("gameMode"),
                "gameName":            info.get("gameName", ""),
                "gameStartTimestamp":  info.get("gameStartTimestamp"),
            })
            # 自动检测 Mayhem queue id
            if is_mayhem(matches[-1]):
                MAYHEM_QUEUE_IDS.add(matches[-1]["queueId"])
    all_matches[puuid] = matches

print(f"✅ Fetched matches for {len(all_matches)} players")
print(f"   Detected Mayhem queue IDs: {MAYHEM_QUEUE_IDS}")

# ── Cell 8: 分析 Mayhem 参与率 ───────────────────────────
results = []
all_puuids = player_df["puuid"].tolist()

for tier in TIERS + ["OVERALL"]:
    if tier == "OVERALL":
        puuids_subset = all_puuids
    else:
        puuids_subset = player_df[player_df["tier"] == tier]["puuid"].tolist()

    n = len(puuids_subset)
    played_mayhem   = 0
    total_games     = 0
    total_mayhem    = 0

    for p in puuids_subset:
        ms = all_matches.get(p, [])
        total_games  += len(ms)
        mc = sum(1 for m in ms if is_mayhem(m))
        total_mayhem += mc
        if mc > 0:
            played_mayhem += 1

    results.append({
        "tier":               tier,
        "n_players":          n,
        "players_w_mayhem":   played_mayhem,
        "player_mayhem_%":    round(played_mayhem / n * 100, 1) if n else 0,
        "total_games":        total_games,
        "mayhem_games":       total_mayhem,
        "game_mayhem_%":      round(total_mayhem / total_games * 100, 1) if total_games else 0,
    })

ratio_df = pd.DataFrame(results)
print("\n📊 ARAM: Mayhem 参与率")
print(ratio_df.to_string(index=False))

# ── Cell 9: 上线前后活跃度对比 ───────────────────────────
from scipy import stats

windowed = {}
for puuid in tqdm(player_df["puuid"], desc="Windowed match counts"):
    pre  = client.get_match_ids(puuid, count=200, start_ts=PRE_START_TS,  end_ts=PRE_END_TS)
    post = client.get_match_ids(puuid, count=200, start_ts=POST_START_TS, end_ts=POST_END_TS)
    windowed[puuid] = {"pre": len(pre), "post": len(post)}

# 分析
activity_results = []
all_pre_counts, all_post_counts = [], []

for tier in TIERS + ["OVERALL"]:
    if tier == "OVERALL":
        subset = all_puuids
    else:
        subset = player_df[player_df["tier"] == tier]["puuid"].tolist()

    pre_counts  = [windowed[p]["pre"]  for p in subset if p in windowed]
    post_counts = [windowed[p]["post"] for p in subset if p in windowed]

    if tier != "OVERALL":
        all_pre_counts.extend(pre_counts)
        all_post_counts.extend(post_counts)

    n = len(pre_counts)
    if n < 2:
        continue

    mean_pre  = sum(pre_counts)  / n
    mean_post = sum(post_counts) / n
    t_stat, p_val = stats.ttest_rel(post_counts, pre_counts)

    activity_results.append({
        "tier":           tier,
        "n":              n,
        "pre_30d_avg":    round(mean_pre, 1),
        "post_30d_avg":   round(mean_post, 1),
        "pre_per_day":    round(mean_pre  / WINDOW_DAYS, 2),
        "post_per_day":   round(mean_post / WINDOW_DAYS, 2),
        "change_%":       round((mean_post - mean_pre) / mean_pre * 100, 1) if mean_pre else None,
        "p_value":        round(p_val, 4),
        "significant":    "✅" if p_val < 0.05 else "❌",
    })

activity_df = pd.DataFrame(activity_results)
print(f"\n📈 Mayhem 上线前后 {WINDOW_DAYS} 天每人日均 LoL 场次")
print(activity_df.to_string(index=False))

# ── Cell 10: 保存结果 ────────────────────────────────────
ratio_df.to_csv("mayhem_ratio.csv", index=False)
activity_df.to_csv("activity_change.csv", index=False)
player_df.to_csv("sampled_players.csv", index=False)

print("\n✅ 结果已保存：mayhem_ratio.csv / activity_change.csv / sampled_players.csv")

# 下载到本地
from google.colab import files
files.download("mayhem_ratio.csv")
files.download("activity_change.csv")
```
